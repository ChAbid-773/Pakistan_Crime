#**Pakistan Crime**

```python
import pandas as pd
import os
```

```python
import kagglehub
path = kagglehub.dataset_download("haqkhan/pakistan-crime-data")
```

```python
crime=pd.read_csv(os.path.join(path,'crimes.csv'))
```

```python
crime
```

```python
crime.info()
```

```python
crime=crime.drop(['Unnamed: 0'],axis=1)
```

```python
crime
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
coder=LabelEncoder()
```

```python
crime['Offence']=crime['Offence'].astype(str)
crime['Offence']=coder.fit_transform(crime['Offence'])
```

```python
crime
```

```python
x=crime.drop(['Offence'],axis=1)
y=crime['Offence']
```

```python
x
```

```python
y
```
